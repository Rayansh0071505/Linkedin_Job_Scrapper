```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException, NoSuchElementException
from selenium.webdriver.common.keys import Keys
import time

# path to the ChromeDriver executable
chromedriver_path = "C:/Users/Amit/Downloads/chromedriver-win64/chromedriver-win64/chromedriver.exe"

s = Service(chromedriver_path)

# List of job search URLs
job_urls = [
    'https://www.linkedin.com/jobs/search?keywords=Machine%20Learning&location=Bengaluru%2C%20Karnataka%2C%20India&geoId=105214831&trk=public_jobs_jobs-search-bar_search-submit&position=1&pageNum=0',
    'https://www.linkedin.com/jobs/search?keywords=Machine%20Learning&location=Austin%2C%20Texas%2C%20United%20States&geoId=104472865&trk=public_jobs_jobs-search-bar_search-submit&position=1&pageNum=0',
    'https://www.linkedin.com/jobs/search?keywords=Machine%20Learning&location=New%20York%2C%20New%20York%2C%20United%20States&geoId=102571732&trk=public_jobs_jobs-search-bar_search-submit&position=1&pageNum=0',
    'https://www.linkedin.com/jobs/search?keywords=Machine%20Learning&location=chennai&geoId=&trk=public_jobs_jobs-search-bar_search-submit&position=1&pageNum=0',
    'https://www.linkedin.com/jobs/search?keywords=Machine%20Learning&location=New%20Delhi%2C%20Delhi%2C%20India&geoId=115918471&trk=public_jobs_jobs-search-bar_search-submit&position=1&pageNum=0',
    'https://www.linkedin.com/jobs/search?keywords=Machine%20Learning&location=Berlin%2C%20Berlin%2C%20Germany&geoId=106967730&trk=public_jobs_jobs-search-bar_search-submit&position=1&pageNum=0',
    'https://www.linkedin.com/jobs/search?keywords=Machine%20Learning&location=Sydney%2C%20New%20South%20Wales%2C%20Australia&geoId=104769905&trk=public_jobs_jobs-search-bar_search-submit&position=1&pageNum=0',
    'https://www.linkedin.com/jobs/search?keywords=Machine%20Learning&location=Waterloo%2C%20Ontario%2C%20Canada&geoId=104043369&trk=public_jobs_jobs-search-bar_search-submit&position=1&pageNum=0',
    'https://www.linkedin.com/jobs/search/?currentJobId=3813136576&geoId=103925994&keywords=Machine%20Learning&location=Tokyo%2C%20Japan&origin=JOB_SEARCH_PAGE_LOCATION_AUTOCOMPLETE&refresh=true',
    'https://www.linkedin.com/jobs/search?keywords=Machine%20Learning&location=Melbourne%2C%20Victoria%2C%20Australia&geoId=100992797&trk=public_jobs_jobs-search-bar_search-submit&position=1&pageNum=0',
    'https://www.linkedin.com/jobs/search?keywords=Machine%20Learning&location=London%2C%20England%2C%20United%20Kingdom&geoId=102257491&trk=public_jobs_jobs-search-bar_search-submit&position=1&pageNum=0',
    'https://www.linkedin.com/jobs/search?keywords=Machine%20Learning&location=Dublin%2C%20County%20Dublin%2C%20Ireland&geoId=105178154&trk=public_jobs_jobs-search-bar_search-submit&position=1&pageNum=0',
]

job_details = []

for job_url in job_urls:
    driver = webdriver.Chrome(service=s)

    try:
        driver.get(job_url)

        WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.XPATH, '//*[@id="main-content"]/section[2]/ul/li')))

        for i in range(50): 
            driver.find_element(By.TAG_NAME, 'body').send_keys(Keys.END)
            time.sleep(2) 

        job_listings = driver.find_elements(By.XPATH, '//*[@id="main-content"]/section[2]/ul/li')

        for listing in job_listings:
            try:
                job_title_element = listing.find_element(By.CLASS_NAME, 'sr-only')
                job_title = job_title_element.text

                address_element = listing.find_element(By.CLASS_NAME, 'base-search-card__metadata')
                address = address_element.text

                link_element = listing.find_element(By.CLASS_NAME, 'base-card__full-link')
                job_link = link_element.get_attribute('href')

                job_details.append({'Job Title': job_title, 'Address': address, 'Job Link': job_link})
            except NoSuchElementException as e:
                print("Element not found:", e)

    except TimeoutException:
        print(f"Timed out waiting for job listings on {job_url}")

    finally:
        driver.quit()

        
df = pd.DataFrame(job_details)

excel_filename = "job_details_scroll_60.xlsx"
df.to_excel(excel_filename, index=False)

print(f"Job details saved to {excel_filename}")

```
